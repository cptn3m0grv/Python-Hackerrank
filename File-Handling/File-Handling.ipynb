```python
# Assume a file city.txt with details of 5 cities in given format (cityname  population(in lakhs) area(in sq KM) ):
# Example:
# Dehradun 5.78 308.20
# Delhi 190 1484
# …………… 
# Open file city.txt and read to:
# Display details of all cities  [3]
# Display city names with population more than 10Lakhs [4]
# Display sum of areas of all cities  [3]
```

```python
l = []
with open("city.txt", "r") as file:
#     data = file.readlines()
    for line in file:
        name, pop, area = line.split()
        pop = float(pop)
        area = float(area)
        l.append([name, pop, area])
        
# print(l)

#display details of all the cities
for name, pop, area in l:
    print(name, pop, area)
    
print()
#display city names with population more than 10 lakhs
for det in l:
    if(det[1]>10):
        print(det[0], det[1])

print()
#display sum of area of all cities
sum = 0
for det in l:
    sum += det[2]
    
print(sum)
```

```python
# Assuming suitable data create a file “temp.txt” which stores the sales of 10 products quarterly in the given format where sales_first is no of sales in the first quarter. Productname sales_first sales_second sales_third sales_fourth
# e.g.  TV 45 78 89 90
#         mobile 123 678 781 772
#         …….
#  Write python script to:
# Display all product details
# Find average sale of all products
# Find product with maximum sales
```

```python
ll = []
with open("temp.txt", "r") as file:
    data = file.readlines()
    for line in data:
        name, q1, q2, q3, q4 = line.split()
        q1, q2, q3, q4 = int(q1), int(q2), int(q3), int(q4)
        ll.append([name, q1, q2, q3, q4])

# Display all product details
for name, q1, q2, q3, q4 in ll:
    print(name, q1, q2, q3, q4)
    
print()
# find average sale of all products
for a in ll:
    av = (a[1]+a[2]+a[3]+a[4])/4
    print(a[0], av)
    
print()    

mxN = 0
name = ""
# fid product with maximum sales
for a in ll:
    prev = mxN
    mxN = max(mxN, (a[1]+a[2]+a[3]+a[4]))
    if(prev != mxN):
        name = a[0]
        
print(name, mxN)
```
